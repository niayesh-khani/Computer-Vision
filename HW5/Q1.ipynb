```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image_path = "./images/q1.png"  # Adjust this to your image path
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Convert the image to binary
# Adjust the threshold value if needed
_, binary_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

# Find connected components
num_labels, labels_im, stats, centroids = cv2.connectedComponentsWithStats(binary_image)

# Define custom colors for each connected component
# Here we include the original set of colors, plus two additional ones
custom_colors = [
    [255, 0, 0],  # Red
    [0, 255, 0],  # Green
    [0, 0, 255],  # Blue
    [255, 255, 0],  # Yellow
    [255, 0, 255],  # Magenta
    [0, 255, 255],  # Cyan
    [138,43,226],  # Purple
    [210,105,30],  # Orange
]

# Initialize the output image with zero values (black)
output_image = np.zeros((binary_image.shape[0], binary_image.shape[1], 3), dtype=np.uint8)

# Assign custom colors to each connected component (excluding the background)
for i in range(1, num_labels):
    # Use the modulo operator to cycle through the list of colors if needed
    color = custom_colors[(i - 1) % len(custom_colors)]
    output_image[labels_im == i] = color

# Display the result with custom colors
plt.imshow(output_image)
plt.title(f"Connected Components")
plt.axis('off')
plt.show()


print("Total number of components:", num_labels - 1)
```
