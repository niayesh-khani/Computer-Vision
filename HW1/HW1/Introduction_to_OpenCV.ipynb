Now try to make Matplotlib output like OpenCV and show it.

```python
##########################Start Code###################################
image = cv2.imread('ComputerDepartment.jpg')

# Convert the image from BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')
plt.show()
###########################End Code####################################
```

Please obtain and display the following modes for color image concatenation.

Mode 1 : The main image should be joined vertically with its vertical invert.

```python
##########################Start Code###################################
image_flipped = np.flipud(image_rgb)

# Concatenate the original image with its flipped version vertically
combined_image = np.concatenate((image_rgb, image_flipped), axis=0)

plt.imshow(combined_image)
plt.axis('off')
plt.show()

###########################End Code####################################
```

Mode 2 : The vertical invert of the original image should be joined vertically with the original image.

```python
##########################Start Code###################################
combined_image = np.concatenate((image_flipped, image_rgb), axis=0)

# Display the combined image using Matplotlib
plt.imshow(combined_image)
plt.axis('off')
plt.show()

###########################End Code####################################
```

Mode 3 : The main image should be joined horizontally with its horizontal invert.

```python
##########################Start Code###################################
image_flipped_horizontal = np.fliplr(image_rgb)

# Concatenate the original image with its horizontally flipped version horizontally
combined_image_horizontal = np.concatenate((image_rgb, image_flipped_horizontal), axis=1)

# Display the combined image using Matplotlib
plt.imshow(combined_image_horizontal)
plt.axis('off')
plt.show()
###########################End Code####################################
```

Mode 4 : The horizontal invert of the original image should be joined horizontally with the original image.

```python
##########################Start Code###################################
combined_image_horizontal = np.concatenate((image_flipped_horizontal, image_rgb), axis=1)

# Display the combined image using Matplotlib
plt.imshow(combined_image_horizontal)
plt.axis('off')
plt.show()
###########################End Code####################################
```

Change the color of one of the windows of the computer faculty building as desired. Try to do this very carefully.

```python
##########################Start Code##################################

# Read the image with OpenCV
image = cv2.imread('ComputerDepartment.jpg')

# Define the coordinates of the rectangle
top_left = (168,225)
bottom_right = (145, 197)

#Set the color and thickness of the rectangle
rectangle_color = (0, 255, 255)
thickness = cv2.FILLED

# Draw the rectangle on the image
cv2.rectangle(image, top_left, bottom_right, rectangle_color, thickness)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')
# plt.show()

###########################End Code####################################
```

First, read and display the image of the iust logo. Then show each of its color channels so that the ​blue channel​ is displayed in the blue, ​green channel​ in the green, and ​red channel​ in red. Notice that since these are color images they have to be ​m​x​n​x​3 ​arrays. For a purely red image, the green and blue channels are entirely zero.

Read logo and show it.

```python
##########################Start Code###################################
logo = cv2.imread('logo.jpg', cv2.IMREAD_UNCHANGED)
logo_rgb = cv2.cvtColor(logo, cv2.COLOR_BGR2RGB)

plt.imshow(logo_rgb)
plt.axis('off')
plt.show()
###########################End Code####################################
```

Show blue channel.

```python
##########################Start Code###################################
blue_channel = logo_rgb.copy()
blue_channel[:, :, 0] = 0  # Set red channel to zero
blue_channel[:, :, 1] = 0  # Set green channel to zero

plt.imshow(blue_channel)
plt.title('Blue Channel')
plt.axis('off')

plt.show()

###########################End Code####################################
```

Show green channel.

```python
##########################Start Code###################################
green_channel = logo_rgb.copy()
green_channel[:, :, 0] = 0  # Set red channel to zero
green_channel[:, :, 2] = 0  # Set blue channel to zero

plt.imshow(green_channel)
plt.title('Green Channel')
plt.axis('off')
plt.show()
###########################End Code####################################
```

Show red channel.

```python
##########################Start Code###################################
red_channel = logo_rgb.copy()
red_channel[:, :, 1] = 0  # Set green channel to zero
red_channel[:, :, 2] = 0  # Set blue channel to zero

plt.imshow(red_channel)
plt.title('Red Channel')
plt.axis('off')

plt.show()
###########################End Code####################################
```
